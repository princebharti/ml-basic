```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
data=pd.read_csv('/Users/princebharti/Desktop/dataset/mnist_train.csv')
label=data['label']
del data['label']
```

```python
scaled_data=StandardScaler().fit_transform(data) # column standardization
model=TSNE(n_components=2,perplexity=20,random_state=0)
n=1500 # no of data points out of 42k points
tsne_data=model.fit_transform(scaled_data[:n])
df=pd.DataFrame(tsne_data,columns=['v1','v2'])
df['label']=label[:n]
# plt.figure(figsize=(6,9))
sns.FacetGrid(data=df,hue='label',size=6).map(plt.scatter,'v1','v2').add_legend()
plt.ylim(-100,100)
plt.xlim(-100,100)
plt.tight_layout()
```

```python
# tsne with perplexity=30 
model=TSNE(n_components=2,perplexity=30,n_iter=5000,random_state=0)
n=1500 # no of data points out of 42k points
tsne_data=model.fit_transform(scaled_data[:n])
df=pd.DataFrame(tsne_data,columns=['v1','v2'])
df['label']=label[:n]
# plt.figure(figsize=(6,9))
sns.FacetGrid(data=df,hue='label',size=6).map(plt.scatter,'v1','v2').add_legend()
plt.ylim(-100,100)
plt.xlim(-100,100)
plt.tight_layout()
```
